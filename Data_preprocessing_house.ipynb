```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/content/independent-house.csv")
```

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
df.head()
```

```python
df.drop(columns = ('link'),inplace = True)
```

```python
df.info()
```

```python
df = df[~df['price'].isna()]
```

```python
df.isna().sum()
```

```python
df.head(3)
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
df['society']=df['society'].fillna('NA')
```

```python
df['society'].replace('nul','NA')
```

```python
#Removing words with different
df['society']=df['society'].str.replace(r'\d+(\.\d+)?\s?★',"",regex = True).str.lower()
```

```python
df.head(2)
```

```python
df = df[df['price']!='Price on Request']
```

```python
def treat_price(x):
  if x[1]=='Lac':
    return round(float(x[0])/100,2)
  else:
    return round(float(x[0]))
```

```python
df['price'] = df['price'].str.split().apply(treat_price)
```

```python
df.head()
```

```python
df.rename(columns = {'rate':'price_per_sqft'},inplace=True)
```

```python
df['price_per_sqft']=df['price_per_sqft'].str.split().str[1].str.split('/').str[0].str.replace(',','').astype('float')
```

```python
df['bedRoom']=df['bedRoom'].str.split().str[0].astype('int')
```

```python
df['bathroom']=df['bathroom'].str.split().str[0].astype('int')
```

```python
df['balcony']=df['balcony'].str.split().str[0]
```

```python
df['balcony'] = df['balcony'].str.split(' ').str.get(0).str.replace('No','0')
```

```python
df['additionalRoom'].fillna('not available',inplace=True)
```

```python
df['additionalRoom'] = df['additionalRoom'].str.lower()
```

```python
df['noOfFloor'].head()
```

```python
df['noOfFloor'].isnull().sum()
```

```python
df['noOfFloor']=df['noOfFloor'].str.split().str[0]
```

```python
df.head()
```

```python
df.rename(columns={'noOfFloor':'floorNum'},inplace=True)
```

```python
df['facing'].fillna('NA',inplace=True)
```

```python
df['area'] = round((df['price']*10000000)/df['price_per_sqft'])
```

```python
df.insert(loc=1,column='property_type',value='house')
```

```python
df.to_csv('house_cleaned.csv',index=False)
```
